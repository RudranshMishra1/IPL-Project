## Loading the Libraries and Dataset

# Here comes some basic questions

```python
# 1. Which team won the most matches?

match_wins = df['match_winner'].value_counts()
sns.barplot(y = match_wins.index, x = match_wins.values,palette='bright')
plt.title('Most match win by teams')
plt.xlabel('Match wins')
plt.ylabel("Team names")
```

```python
# 2. Toss decision trends

sns.countplot(x=df['toss_decision'], palette='viridis')
plt.title("Toss Decision Trends")
```

```python
# Toss winner VS Match winner

count = df[df['toss_winner'] == df['match_winner']]['match_id'].count()
percentage = (count*100)/df.shape[0]
percentage.round(1)
```

```python
# How do teams win? (Runs vs Wickets)
sns.countplot(x=df['won_by'])
plt.title('Won By')
```

```python
# Player of the match awards
count = df['player_of_the_match'].value_counts().head(10)
sns.barplot(x=count.values,y=count.index,palette='mako')
```

```python
# 2 top scorers
high = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(2)
high.plot(kind='barh')
```

```python
# 10 best bowling figures
df.head()
```

```python
df['highest_wickets'] = df['best_bowling_figure'].apply(lambda x :x.split('--')[0])
df['highest_wickets'] = df['highest_wickets'].astype(int)
wick = df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending=False).head(10)
wick.plot(kind='barh')
```

```python
# Most matches player by venue
venue = df['venue'].value_counts()
```

```python
# Who won the highest margin by runs?
df[df['won_by'] == 'Runs'].sort_values(by='margin',ascending=False).head(1)[['match_winner','margin']]
```

```python
df.nlargest(1,'margin')[['match_winner','margin']]
```

```python
# which bowler had the best bowling figures?
df.nlargest(4, 'highest_wickets')[['best_bowling','best_bowling_figure']]
```
